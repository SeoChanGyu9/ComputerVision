```python
gray_blur = cv2.GaussianBlur(gray, (15, 15), 0)
#gray_blur = cv2.medianBlur(gray,5) #그림자없애줌
#gray_blur = cv2.GaussianBlur(gray_blur, (15, 15), 0)

#테두리흰색
thresh = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV, 11, 1)
#테두리검은색
#thresh = cv2.adaptiveThreshold(gray_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
cv2.imshow('gray',gray)
cv2.imshow('gray_blur',gray_blur)
cv2.imshow('thresh',thresh)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((3, 3), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=4)

cv2.imshow('thresh',thresh)
cv2.imshow('closing',closing)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
def run_main():
    img = cv2.imread('h3test.png',cv2.IMREAD_COLOR)
    #img = cv2.imread('sIMG_8253.jpg',cv2.IMREAD_COLOR)
    img = cv2.resize(img, dsize=(494,492), interpolation=cv2.INTER_AREA)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    while(True):
        gray_blur = cv2.GaussianBlur(gray, (15, 15), 0)
        #thresh = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV, 11, 1)
        thresh = cv2.adaptiveThreshold(gray_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
        kernel = np.ones((3, 3), np.uint8)
        closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=4)
        cont_img = closing.copy()
        contours, hierarchy = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,
                                               cv2.CHAIN_APPROX_SIMPLE)
        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area < 2000 or area > 4000:
                continue
            if len(cnt) < 5:
                continue
            ellipse = cv2.fitEllipse(cnt)
            cv2.ellipse(img, ellipse, (0,255,0), 2)
        cv2.imshow("Morphological Closing", closing)
        cv2.imshow("Adaptive Thresholding", thresh)
        cv2.imshow('Contours', img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cv2.destroyAllWindows()
if __name__ == "__main__":
    run_main()
```

```python
import numpy as np
import cv2


#img = cv2.imread('h3test.png') # 컬러
img = cv2.imread('h3test2.png') # 컬러
#img = cv2.imread('sIMG_8253.jpg',)

img = cv2.resize(img, dsize=(494,492), interpolation=cv2.INTER_AREA)
img_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # 흑백

img_g_blur = cv2.GaussianBlur(img_g, (7, 7), 0)
img_thresh = cv2.adaptiveThreshold(img_g_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 45, 6)



kernel = np.ones((3, 3), np.uint8)
#closing = cv2.morphologyEx(img_thresh, cv2.MORPH_CLOSE, kernel, iterations = 9)
closing = cv2.morphologyEx(img_thresh, cv2.MORPH_CLOSE, kernel, iterations = 20)




cont_img = closing.copy()
contours, _ = cv2.findContours(cont_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours :
    area = cv2.contourArea(cnt)
    if area < 1000 or area > 4000 : continue
    if len(cnt) < 5 : continue
    
    ellipse = cv2.fitEllipse(cnt)
    cv2.ellipse(img, ellipse, (0, 255, 0), 2)

cv2.imshow('img_thresh',img_thresh)
cv2.imshow('cont_img',cont_img)
cv2.imshow('img',img)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def nothing(x):
    pass



cv2.namedWindow('image')

cv2.createTrackbar('hue','image',0,360,nothing)
cv2.createTrackbar('S','image',0,100,nothing)
cv2.createTrackbar('V','image',0,100,nothing)

#                   트랙바이름   윈도우이름  트랙바범위  ?
img = cv2.imread('sIMG_8253.jpg')


while(1):
    hue = cv2.getTrackbarPos('hue','image')
    S = cv2.getTrackbarPos('S','image')
    V = cv2.getTrackbarPos('V','image')
    
    lower_blue = np.array([hue-50, S-50, V-50])
    upper_blue = np.array([hue+50, S+50, V+50])

    mask = cv2.inRange(img, lower_blue, upper_blue)

    res = cv2.bitwise_and(img,img, mask= mask)
    
    cv2.imshow('image',img)
    cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    k = cv2.waitKey(1) & 0xFF # esc종료
    if k == 27:
        break

cv2.destroyAllWindows()
```
